```python
numeric = df.select_dtypes(include=["int64", "float64"])

for column in numeric:
    z_score = (numeric[column] - numeric[column].mean())/numeric[column].std()
    outliers = df[np.abs(z_score) > 3]
    print(column)
    if not outliers.empty:
      print(outliers[column])
    else:
      print("Нет выбросов")
    print()
```

##3.

```python
sns.histplot(df["Previous Purchases"], bins=20, kde=True)
plt.title("Частота количества прошлых покупок")
plt.xlabel("Количество прошлых покупок")
plt.ylabel("Частота")
plt.show()
```

```python
rating_trend = df.groupby("Previous Purchases")["Review Rating"].mean()

plt.plot(rating_trend)
plt.title("Средний рейтинг в зависимости от количества прошлых покупок")
plt.xlabel("Количество прошлых покупок")
plt.ylabel("Средний рейтинг")
plt.show()
```

```python
df["Payment Method"].value_counts().plot(kind="bar")
plt.title("Способы оплаты")
plt.xlabel("Способ оплаты")
plt.ylabel("Количество")
plt.show()
```

```python
sns.boxplot(x=df["Gender"], y=df["Purchase Amount (USD)"])
plt.title("Сумма покупок по полу")
plt.xlabel("Пол")
plt.ylabel("Сумма покупки (USD)")
plt.show()
```

```python
sns.heatmap(c, annot=True, cmap="OrRd")
plt.show()
```

##4.

```python
location_stats = df.groupby("Location").agg({"Review Rating": "mean"})
print("Средний рейтинг по локациям:")
print(location_stats)
```

```python
prev_purchase_stats = df.groupby("Previous Purchases").agg({"Customer ID": "count"})
print("Количество клиентов по количеству прошлых покупок:")
print(prev_purchase_stats)
```

```python
freq_stats = df.groupby("Frequency of Purchases")["Purchase Amount (USD)"].agg(["min", "mean", "max"])
print("Статистика суммы покупки по частоте покупок:")
print(freq_stats)
```
